## Module1
- Python example
- Rotate image

```python
import os
import cv2
import ipywidgets as widgets
from IPython.display import display, Image, clear_output

def process(input_path, output_path, option):
    image = cv2.imread(input_path)
    if (option.lower() == "right"):
        rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
    else:
        rotated_image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
    cv2.imwrite(output_path, rotated_image)

def run_module1(output_widget, input_path, output_path, option):
    process(input_path, output_path, option)

    with output_widget:
        clear_output(wait=True)
        print(f"[Module1] Complete: {option}")

def show_result(output_widget, file_path):
    _, ext = os.path.splitext(file_path)
    img = cv2.imread(file_path)
    _, buf = cv2.imencode(ext, img)

    with output_widget:
        clear_output(wait=True)
        print(f"[Module1] {file_path}")
        display(Image(data=buf.tobytes(), width=400))

def update_default_output_path(change):
    dropdown_value = change["new"]
    current_dir = os.getcwd()
    default_output_path = os.path.join(current_dir, f"data/Lenna_{dropdown_value}.png")
    form.children[1].value = default_output_path

def create_form():
    current_dir = os.getcwd()

    # create dropdown menu
    dropdown = widgets.Dropdown(
        options=["right", "left"],
        value="right",
        description="Option:",
    )
    dropdown.layout.width = "200px"
    # observe dropdown value change
    dropdown.observe(update_default_output_path, names="value")

    # create text field
    text1 = widgets.Text(description="Input Path:", 
                         value=os.path.join(current_dir, "data/Lenna.png"))
    text2 = widgets.Text(description="Output Path:", 
                         value=os.path.join(current_dir, f"data/Lenna_{dropdown.value}.png"))
    text1.layout.width = "600px"
    text2.layout.width = "600px"

    # create output area
    output_widget1 = widgets.Output()
    output_widget2 = widgets.Output()

    # create button1
    button1 = widgets.Button(description="Run Module1")
    button1.on_click(lambda _: run_module1(output_widget1, text1.value, text2.value, dropdown.value))

    # create button2
    button2 = widgets.Button(description="Show Result")
    button2.on_click(lambda _: show_result(output_widget2, text2.value))

    form_items = [
        text1,
        text2,
        dropdown,
        button1,
        button2,
        output_widget1,
        output_widget2
    ]
    form = widgets.VBox(form_items, layout=widgets.Layout(width="800px"))

    return form

form = create_form()
display(form)
```
